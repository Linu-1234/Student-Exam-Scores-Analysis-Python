```python
#Deleting column
df.drop('Unnamed: 0',axis=1,inplace=True)
```

# Data Analysis

#### Gender Distribution

```python
plt.figure(figsize=(4,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title('Gender Distribution')
plt.show()
```

* From the above chart , we have analyzed that the number of females in the data is more than the number of males.

#### Parent Education Impact

```python
gb = df.groupby('ParentEduc').agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
print(gb)
```

```python
plt.figure(figsize=(4, 4))
sns.heatmap(gb)
plt.title('Relationship between parents education and students score')
plt.show()
```

* From the above chart , We have concluded that the eduaction of the parents have a good impact on the students' score.

#### Marital Status Impact

```python
gb1=df.groupby('ParentMaritalStatus').agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
print(gb1)
```

```python
plt.figure(figsize=(4, 4))
sns.heatmap(gb1)
plt.title('Relationship between parents marital status and students score')
plt.show()
```

* From the above chart , We have concluded that the marital status of the parents have negligible impact on the students' score.

#### Subject comparision

```python
sns.boxplot(data=df,x='ReadingScore')
plt.show()
```

```python
sns.boxplot(data=df,x='WritingScore')
plt.show()
```

#### Ethnic Groups Distribution

```python
# Extracing unique values from EthnicGroup Column
print(df['EthnicGroup'].unique())
```

```python
groupA = df.loc[(df['EthnicGroup'] == 'group A')].count()
groupB = df.loc[(df['EthnicGroup'] == 'group B')].count()
groupC = df.loc[(df['EthnicGroup'] == 'group C')].count()
groupD = df.loc[(df['EthnicGroup'] == 'group D')].count()
groupE = df.loc[(df['EthnicGroup'] == 'group E')].count()
l=['groupA','groupB','groupC','groupD','groupE']
mlist=[groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]
plt.pie(mlist,labels=l,autopct='%1.2f%%')
plt.title('Distribution of Ethnic Groups')
plt.show()
```

* From the above pie chart ,We have concluded that Ethnic group C  has more count than other groups.
